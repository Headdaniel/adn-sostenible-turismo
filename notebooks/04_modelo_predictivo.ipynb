### Interpretación honesta de las métricasEl R² de test (0.9979) es **artificialmente alto** y no debe leerse como "el modelo prediceel futuro con 99.8% de exactitud".**Razón:** el target (`score_pca`) es una combinación lineal (PCA) de las mismas variablesque alimentan el modelo. El XGBoost, por tanto, aprende a **reproducir una fórmula**, no apredecir un fenómeno externo no observado.**Valor real del modelo:** funciona como un **motor de inferencia** que permite recalcular elscore ante nuevas combinaciones de variables (simulación de escenarios, proyecciones,zonas sin score previo), y que se sirve en producción vía API.La validación cruzada (R²=0.9606 ± 0.018) es una medida más conservadora y realista dela estabilidad del modelo.
